```python
import os
import pandas as pd

DATA_DIR = "data"

texts = []
file_names = []

for fname in sorted(os.listdir(DATA_DIR)):
    if fname.endswith(".txt"):
        with open(os.path.join(DATA_DIR, fname), "r", encoding="utf-8") as f:
            texts.append(f.read())
        file_names.append(fname)

df = pd.DataFrame({
    "Filename": file_names,
    "Document": texts
})

df.head()
```

```python
df["Text"] = df["Document"].str.replace(r"\s+", " ", regex=True).str.strip()
df[["Filename", "Document", "Text"]].head()
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")  # use a different model if not English

def annotate(text):
    doc = nlp(text)
    tokens = [t.text for t in doc if not t.is_space]
    lemmas = [t.lemma_ for t in doc if not t.is_space]
    pos = [t.pos_ for t in doc if not t.is_space]
    return " ".join(tokens), " ".join(lemmas), " ".join(pos)

df[["Tokens", "Lemmas", "Parts-of-speech"]] = df["Text"].apply(
    lambda x: pd.Series(annotate(x))
)

df[["Filename", "Tokens", "Lemmas", "Parts-of-speech"]].head()

```
